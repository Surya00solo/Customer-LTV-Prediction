### Subtask:
Merge customer data with transaction history.


**Reasoning**:
Create simulated transaction data and merge it with the customer data.



```
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 1. Create simulated transaction data
transactions = []
for index, row in df_cleaned.iterrows():
    customer_id = row['customer_id']
    num_purchases = int(row['total_purchases'])
    avg_purchase_value = row['avg_purchase_value']
    for _ in range(num_purchases):
        transaction_value = np.random.normal(loc=avg_purchase_value, scale=avg_purchase_value * 0.2)
        # Ensure transaction value is not negative
        transaction_value = max(0, transaction_value)
        days_ago = np.random.randint(0, 365)  # Transactions within the last year
        transaction_date = (datetime.now() - timedelta(days=days_ago)).strftime('%Y-%m-%d')
        transactions.append([customer_id, transaction_value, transaction_date])

df_transactions = pd.DataFrame(transactions, columns=['customer_id', 'transaction_value', 'transaction_date'])

# 2. Merge the dataframes
df_merged = pd.merge(df_cleaned, df_transactions, on='customer_id', how='left')
display(df_merged.head())
```

## Feature engineering

### Subtask:

**Reasoning**:
Calculate Frequency, Recency, AOV, and Total Spend for each customer and handle the data types.



```
# Convert transaction_date to datetime objects
df_merged['transaction_date'] = pd.to_datetime(df_merged['transaction_date'])

# Calculate Frequency (total number of transactions per customer)
frequency_df = df_merged.groupby('customer_id').size().reset_index(name='frequency')

# Calculate Recency (days since last transaction for each customer)
# Define today's date for recency calculation
today = datetime.now()
recency_df = df_merged.groupby('customer_id')['transaction_date'].max().reset_index(name='last_transaction_date')
recency_df['recency'] = (today - recency_df['last_transaction_date']).dt.days
recency_df = recency_df[['customer_id', 'recency']]

# Calculate Monetary Value (AOV - Average Order Value)
aov_df = df_merged.groupby('customer_id')['transaction_value'].mean().reset_index(name='aov')

# Calculate Total Spend
total_spend_df = df_merged.groupby('customer_id')['transaction_value'].sum().reset_index(name='total_spend')

# Merge the calculated features
df_features = frequency_df.merge(recency_df, on='customer_id', how='left')
df_features = df_features.merge(aov_df, on='customer_id', how='left')
df_features = df_features.merge(total_spend_df, on='customer_id', how='left')

display(df_features.head())
```

## Model training

Train an XGBoost regression model.

Import the necessary library, initialize and train the XGBoost regressor model, make predictions on the validation set, and evaluate the model using MAE and RMSE.



```
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Initialize an XGBoost regressor model
# Set a random state for reproducibility
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Train the XGBoost model using the training data
xgb_model.fit(X_train, y_train)

# Make predictions on the validation set
y_pred = xgb_model.predict(X_val)

# Evaluate the model's performance using MAE and RMSE
mae = mean_absolute_error(y_val, y_pred)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))

# Print the evaluation metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

## Model optimization

### Subtask:
Optimize the XGBoost model's hyperparameters.


**Reasoning**:
Define the parameter grid and use GridSearchCV to find the best hyperparameters for the XGBoost model.



```
from sklearn.model_selection import GridSearchCV
import xgboost as xgb

# Define the parameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# Initialize the XGBoost regressor
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)

# Initialize GridSearchCV
# Using neg_mean_absolute_error as the scoring metric, higher is better
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid,
                           scoring='neg_mean_absolute_error', cv=3, n_jobs=-1, verbose=2)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Extract the best hyperparameters and best estimator
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_
best_score = grid_search.best_score_

print("Best hyperparameters:", best_params)
print("Best negative MAE (from GridSearchCV):", best_score)
```

## Data clustering

### Subtask:
Segment customers into clusters based on their predicted LTV.


**Reasoning**:
Segment customers into clusters based on their predicted LTV using KMeans clustering, and then analyze the characteristics of each cluster.



```
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Use the features for clustering
features_for_clustering = ['frequency', 'recency', 'aov', 'total_spend', 'avg_purchase_interval']
X_cluster = df_features[features_for_clustering]

# Determine the optimal number of clusters (e.g., using the elbow method)
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_cluster)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

# Based on the elbow method, choose an appropriate number of clusters (e.g., 3)
optimal_k = 3  # Replace with the actual k value from the elbow method

# Perform KMeans clustering with the chosen k
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_features['cluster'] = kmeans.fit_predict(X_cluster)

# Analyze the characteristics of each cluster
cluster_analysis = df_features.groupby('cluster')[features_for_clustering].agg(['mean', 'std'])
display(cluster_analysis)

# Visualize the clusters (example: using the first two features)
plt.figure(figsize=(8, 6))
for cluster in range(optimal_k):
    plt.scatter(df_features[df_features['cluster'] == cluster]['frequency'],
                df_features[df_features['cluster'] == cluster]['recency'],
                label=f'Cluster {cluster}')
plt.xlabel('Frequency')
plt.ylabel('Recency')
plt.title('Customer Clusters')
plt.legend()
plt.show()
```
